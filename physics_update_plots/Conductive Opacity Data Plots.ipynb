```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
plt.matplotlib.rcParams['figure.figsize'] = (15,15)
plt.matplotlib.rcParams['font.size'] = 30
```

```python
def oneorthree(name):
    try:
        return float(name[8:12])
    except:
        return float(name[8:10])
def fehplot(mass):
    os.chdir('/home/zookeeper/Desktop/workdir/compare01.14/Compare')
    filename = os.listdir()
    oldLTRGB = np.zeros((6,3))
    newLTRGB = np.zeros((6,3))
    oldHeCM = np.zeros((6,3))
    newHeCM = np.zeros((6,3))
    oldidxLTRGB = 0
    newidxLTRGB = 0
    for file in filename:
        if file[1:5]==str(mass):
            if file[-5]=='9':
                dp = pd.read_csv(file)
                oldLTRGB[oldidxLTRGB,0] = oneorthree(file)
                oldHeCM[oldidxLTRGB,0] = oneorthree(file)
                i = 0
                maxi = len(dp['Luminosity: triple-alpha'].values)
                while dp['Luminosity: triple-alpha'].values[i] <= 1e30 and i < maxi-1:
                    i += 1
                oldLTRGB[oldidxLTRGB,1] = dp['log(L/Lsun)'].values[i]
                oldHeCM[oldidxLTRGB,1] = dp['M He core'].values[i]
                if file[10:12] == 'af':
                    oldLTRGB[oldidxLTRGB,2] = float(file[9:10])
                    oldHeCM[oldidxLTRGB,2] = float(file[9:10])
                else:
                    oldLTRGB[oldidxLTRGB,2] = float(file[9:12])
                    oldHeCM[oldidxLTRGB,2] = float(file[9:12])
                oldidxLTRGB += 1
            else:
                dp = pd.read_csv(file)
                newLTRGB[newidxLTRGB,0] = oneorthree(file)
                newHeCM[newidxLTRGB,0] = oneorthree(file)
                i = 0
                maxi = len(dp['Luminosity: triple-alpha'].values)
                while dp['Luminosity: triple-alpha'].values[i] <= 1e30 and i < maxi-1:
                    i += 1
                newLTRGB[newidxLTRGB,1] = dp['log(L/Lsun)'].values[i]
                newHeCM[newidxLTRGB,1] = dp['M He core'].values[i]
                if file[10:12] == 'af':
                    newLTRGB[newidxLTRGB,2] = float(file[9:10])
                    newHeCM[newidxLTRGB,2] = float(file[9:10])
                else:
                    newLTRGB[newidxLTRGB,2] = float(file[9:12])
                    newHeCM[newidxLTRGB,2] = float(file[9:12])
                newidxLTRGB += 1
    oldLTRGB=sorted(oldLTRGB,key=lambda x:x[2])
    newLTRGB=sorted(newLTRGB,key=lambda x:x[2])
    oldHeCM=sorted(oldHeCM,key=lambda x:x[2])
    newHeCM=sorted(newHeCM,key=lambda x:x[2])
    oldxLTRGB = np.zeros(6)
    oldyLTRGB = np.zeros(6)
    newxLTRGB = np.zeros(6)
    newyLTRGB = np.zeros(6)
    oldyHeCM = np.zeros(6)
    newyHeCM = np.zeros(6)
    diffLTRGB = np.zeros(6)
    diffHeCM = np.zeros(6)
    for i in range(6):
        oldxLTRGB[i] = oldLTRGB[i][0]
        oldyLTRGB[i] = oldLTRGB[i][1]
        newxLTRGB[i] = newLTRGB[i][0]
        newyLTRGB[i] = newLTRGB[i][1]
        oldyHeCM[i] = oldHeCM[i][1]
        newyHeCM[i] = newHeCM[i][1]
        diffHeCM[i] = (newyHeCM[i] - oldyHeCM[i])/oldyHeCM[i]
        diffLTRGB[i] = (newyLTRGB[i] - oldyLTRGB[i])/oldyLTRGB[i]
    plt.plot(newxLTRGB,diffLTRGB,'--o',label='$\Delta \log{(L/L_{\odot})}TRBG$')
    plt.plot(newxLTRGB,diffHeCM,'-o',label='$\Delta He Core Mass$')
    plt.xlabel('[Fe/H]')
    plt.title('M = {}'.format(mass))
    plt.tick_params(direction='in',length=20,width=2,bottom=True,top=True,left=True,right=True)
    plt.legend();
```
