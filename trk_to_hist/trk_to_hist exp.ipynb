```python
os.chdir('E:\\share2\\untar')
filestotal = os.listdir()
totalfile = len(filestotal)
while totalfile > 0:
    filesnow = filestotal[0:7]
    print(filesnow)
    pool = Pool(7)
    pool.map(trk_to_hist.convert,filesnow)
    time.sleep(2)
    gc.collect()
    print("closing  pool")
    pool.close()
    print("closed pool")
    print("joining pool")
    pool.join()
    print("joined pool")
    time.sleep(1)
    filestotal = filestotal[7:]
```
